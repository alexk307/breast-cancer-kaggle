```python
from sklearn.model_selection import train_test_split
```

```python
data.corr()
```

Split training data and test data

```python
training_set, test_set = train_test_split(data, test_size=0.2, random_state=60)
```

```python
test_set_labels = test_set.diagnosis.tolist()
```

```python
train_set_labels = training_set.diagnosis.tolist()
```

```python
del test_set['diagnosis']
```

```python
del training_set['diagnosis']
```

```python
test_set_data = test_set.values.tolist()
training_set_data = training_set.values.tolist()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipe = Pipeline([
    ('scaler', StandardScaler())
])

test_set_data = pipe.fit_transform(test_set_data)
training_set_data = pipe.fit_transform(training_set_data)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
```

```python
clf.fit(training_set_data, train_set_labels)
```

```python
for i in range(len(test_set_data)):
    print 'predicted: %s, actual: %s' % (clf.predict(test_set_data[i]), test_set_labels[i])
```
